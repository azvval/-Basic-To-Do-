```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 3. kod bloğudur.
#Bu kod bloğu kullanıcının görevlerinin listelenmesi için bir fonksiyon.
def gorev_liste():
      if not gorev_listesi:
       print("*---*Görev Listesi*---*", "\n")
       print("Görev listeniz boş.")
      else:
        print("\n","*---*Görev Listesi*---*", "\n")
        for sıra, öğe in enumerate(gorev_listesi, 1): # Görevleri kullanıcı girişine göre sıralamak için bir döngü ve enumerate fonskiyonu kullanıyorum.
         print("{}. {}".format(sıra, öğe),"\n")

      gorev_kaydet()
```

```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 4. kod bloğudur.
#Bu kod bloğu kullanıcının girdiği görevleri düzenleyebilmesi için gerekli fonksiyondur.
def gorev_duzenle():
   try:
       duzenle=int(input("Düzenlemek istediğiniz görev numarasını giriniz : "))
       if 1<=duzenle<=len(gorev_listesi):
             duzenle=duzenle-1
             yeni_gorev=input("Yeni görev giriniz : ")
             gorev_listesi[duzenle] = yeni_gorev
             print("Görev güncellendi.")
       else:
            print("Görev bulunamadı. ")
   except ValueError:
       print("Lütfen geçerli bir sayı girin.")

   gorev_kaydet()
```

```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 5. kod bloğudur.
#Bu kod bloğu kullanıcının girdiği görevleri silmesini sağlayan fonksiyonu içerir.
def gorev_sil():
      if not gorev_listesi:
       print("Görev listeniz boş.")
      else:
       gorev_sil=int(input("Silmek istediğiniz görevin sırasını giriniz : ")) - 1
       if 0 <= gorev_sil < len(gorev_listesi):
         silinen_gorev = gorev_listesi.pop(gorev_sil)
         print(f"'{silinen_gorev}' görevi silindi.")
       else:
         print("Geçersiz görev numarası.")
      gorev_kaydet()
```

```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 6. kod bloğudur.
#Bu kod bloğu kullanıcının yeni görevler eklemesini sağlayan fonksiyonu içerir.
def gorev_ekle():
      gorev =input(" Görev : ")
      if len(gorev)==0:
        print("Görev kısmı boş bırakılamaz")
      else:
       gorev_listesi.append(gorev)
       print("Görev eklendi.")
       gorev_kaydet()
```

```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 7. kod bloğudur.
#Her adımda yapılan işlemi bir metin dosyasına kaydetmeyi sağlayan fonksiyonu içerir.
def gorev_kaydet():
  dosya_adi = "TO-DO.txt"
  with open(dosya_adi, 'w', encoding='utf-8') as f:
    for gorev in gorev_listesi:
      f.write(gorev + '\n')
  print("Görevler dosyaya kaydedildi.")
```

```python
#1. kod bloğunu çalıştırmadan önce bu bloğu bir kez çalıştırın lütfen.
#Bu 8. kod bloğudur.
# Daha önce kaydedilen bir görev listesi varsa bulup okuyan ve kullanıcıya gösteren fonksiyonu içerir.
def gorevleri_yukle():
  global gorev_listesi
  dosya_adi = "TO-DO.txt"
  try:
    with open(dosya_adi, 'r', encoding='utf-8') as f:
      gorev_listesi = [satir.strip() for satir in f]
    print("Görevler dosyadan yüklendi.")
  except FileNotFoundError:
    gorev_listesi = []
    print("Görev dosyası bulunamadı, boş bir liste ile başlanıyor.")
  except Exception as e:
    gorev_listesi = []
    print(f"Dosya yüklenirken bir hata oluştu: {e}, boş bir liste ile başlanıyor.")
```
